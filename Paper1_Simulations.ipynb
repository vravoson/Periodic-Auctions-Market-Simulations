# 1. Fixed number of market participants at each auction (Fricke & Gerig Model, 2018)

 ## 1.1 Model

 We consider a single security traded in a periodic auction market. The unobservable price evolves as a brownian motion with variance $\frac 32\psi^2$, i.e:
 $$\forall t\ge 0,\quad m_t = \sqrt{\frac 32}B_t.$$
 We assume that traders arrive to the auction with a fixed frequency $\omega$, so that if the auction lasts $\tau$ seconds, there are $K=\omega\tau$ traders during the auction. We assume that the traders arrive at regular time intervals, i.e. every $\frac{\tau}{K}=\frac{1}{\omega}$ seconds. Everytime a trader $i\in\{1,...,K\}$ comes to the auction, they submit a resevation price $r_{\frac{i}{\omega}}$ (= their estimation of the market price at time $i/\omega$), defined by:
 $$r_{\frac{i}{\omega}} = m_{\frac{i}{\omega}} + g_i,$$
 where $g_1,...,g_K$ are iid $\mathcal{N}(0,\sigma^2)$ that represent the market participants uncertainty on what security price. We also assume that the $g_i$'s are independent from the 'unobservable prices' $(m_t)_{t\ge 0}$
 
 Moreover, in each auction, every trader submits a demand schedule which is linear and decreasing in the auction clearing price:
 $$\forall i\in\{1,...,K\},\quad D_i(p):= A\left(r_{\frac{i}{\omega}}-p\right),$$
 which means that if the auction clearing price is $\overline{p}$ and $D_i(\overline{p})>0$, the $i$-th trader buys $D_i(\overline{p})$ securities at price $\overline{p}$, and it sells the opposite quantity if $D_i(\overline{p})<0$. We assume that $A$ (= volume/price elasticity) is constant for all the traders.
 
 Then, the auction clearing price $\overline{p}$ is the one that generates no excess demand, i.e. the price such that (volume of buyers = volume of sellers), i.e.
 $$\sum_{i\,|\, A(r_{\frac{i}{\omega}}-\overline{p})>0} A\left(r_{\frac{i}{\omega}}-\overline{p}\right) = \sum_{i\,|\, A(r_{\frac{i}{\omega}}-\overline{p})<0} -A\left(r_{\frac{i}{\omega}}-\overline{p}\right),$$
 i.e.
 $$\sum_{i=1}^K A\left(r_{\frac{i}{\omega}}-\overline{p}\right) =0$$
 i.e.
 $$ \overline{p} = \frac{1}{K}\sum_{i=1}^K r_{\frac{i}{\omega}}$$

 The market inefficiency metric, called the $\textit{liquidity risk}$, is defined as follows:
 $$L_1(\tau):=\frac{1}{K}\sum_{i=1}^K\text{Var}\left(\overline{p} - m_{\frac{i}{\omega}} \right).$$

 $\textbf{[Not in the paper]}$ - For $i\in\{1,...,K\}$, we have:

 $$\text{Var}\left(\overline{p} - m_{\frac{i}{\omega}} \right) = \text{Var}\left(\frac{1}{K}\sum_{k=1}^K \left(g_k + r_{\frac{k}{\omega}}\right) - m_{\frac{i}{\omega}} \right)= \frac{\sigma^2}{K}+\text{Var}\left(\frac{1}{K}\sum_{k=1}^K\left( m_{\frac{k}{\omega}} - m_{\frac{i}{\omega}} \right) \right) $$

 We notice that:

 $$\text{Var}\left(\frac{1}{K}\sum_{k=1}^K\left( m_{\frac{k}{\omega}} - m_{\frac{i}{\omega}} \right) \right) = \text{Var}\left(\sqrt{\frac 32}\frac{\psi}{K}\sum_{k=1}^K\left( B_{\frac{k}{\omega}} - B_{\frac{i}{\omega}} \right) \right) = \frac{3}{2}\frac{\psi^2}{K^2}\text{Var}\left(\sum_{k=1}^{i-1}k\left(B_{\frac{k}{\omega}}-B_{\frac{k+1}{\omega}} \right)+\sum_{k=i+1}^{K}(K-k+1)\left(B_{\frac{k}{\omega}}-B_{\frac{k-1}{\omega}} \right) \right)=\frac{3}{2}\tau\frac{\psi^2}{K^3}\left(\sum_{k=1}^K k^2+\sum_{k=1}^{K-i}k^2 \right)$$
 After computing the sums, we get:
 $$\text{Var}\left(\overline{p} - m_{\frac{i}{\omega}} \right) = \frac{\tau\psi^2}{2}\left[1+\frac{3}{2K}+\frac{1}{2K^2}-\frac{3i}{K}-\frac{3i}{K^2}+\frac{3i^2}{K^2}\right] $$

 So:
 $$L_1(\tau)=\frac{1}{K}\sum_{i=1}^K\text{Var}\left(\overline{p} - m_{\frac{i}{\omega}} \right) = \frac{\sigma^2}{K}+\frac{\tau\psi^2}{2}\left[1+\frac{3}{2K}+\frac{1}{2K^2}-\frac{3}{2}\left(1+\frac{1}{K}\right)-\frac{3(K+1)}{2K^2}+\frac{K(K+1)(2K+1)}{2K^3} \right] = \left(\frac{\sigma^2}{\omega} - \frac{\psi^2}{4\omega^2} \right)\frac{1}{\omega}+\frac{\psi^2}{4}\tau.$$

 If we assume that there is at least one trader during the auction (i.e. $K=\omega\tau\ge 1$), then the optimal auction duration is the one that minimizes the market inefficiency:
 $$\tau^{*}:=\underset{\tau\ge \frac{1}{\omega}}{\text{Argmin } L_1(\tau)}$$
 From the equation above, we find:
 $$\tau^{*}:=\sqrt{\frac{4\sigma^2}{\omega\psi^2} - \frac{1}{\omega^2}},$$
 which is well defined if $\omega > \psi^2/(4\sigma^2)$. In the paper, the authors make the implicit assumption that there is a high number of traders, i.e.
 $$ \omega >> \frac{\sigma^2}{\psi^2},$$
 which leads to the following liquidity risk:
 $$L_1(\tau) \approx \frac{\sigma^2}{\omega\tau} + \frac{\tau\psi^2}{4} $$
 and
 $$\tau^{*}\approx \frac{2\sigma}{\psi\sqrt{\omega}}. $$

 ## 1.2 Agent-based simulation

```python
import numpy as np
import tqdm
import matplotlib.pyplot as plt

def get_liquidity_risk(tau, omega, psi, sigma, n_simulations = 100):
    """
    This function simulates the liquidity risk n_simulations times for every timestep tau
    and average the results.
    """
    liquidity_risk_samples = []
    K = int(omega*tau)
    if K==0:
        return 100
    else:
        for sim in range(n_simulations):
            traders_arrival_times = 1/omega*np.arange(1,K+1)
            unobservable_price = np.cumsum(np.random.normal(0,psi/np.sqrt(omega),K))
            reservation_prices = unobservable_price + np.random.normal(0, sigma, K)
            clearing_price = np.mean(reservation_prices)
            liquidity_risk_sample = clearing_price*np.ones(K) - unobservable_price
            liquidity_risk_samples.append(liquidity_risk_sample)
        liquidity_risk_list = list(map(np.var, np.transpose(liquidity_risk_samples).tolist()))
        return np.mean(liquidity_risk_list)
```

```python
# PARAMETERS
omega = 30
psi = 2
sigma = 1
tau_vals = np.linspace(0.001,2,1000)
liquidity_risks = []
for tau in tqdm.tqdm(tau_vals):
    liquidity_risk_tau = get_liquidity_risk(tau, omega, psi, sigma)
    liquidity_risks.append(liquidity_risk_tau)
    
plt.plot(tau_vals, liquidity_risks, "b", label = "Agent-Based Simulation")
plt.plot(tau_vals, sigma**2/(omega*tau_vals)+1/4*2/3*psi**2*tau_vals, 'r--', label = "Theoretical Result") #Theoretical result
plt.xlabel(r'Auction Duration $\tau$')
plt.ylabel(r'Liquidity Risk')
plt.legend(loc = "best")
plt.ylim([0,1])
plt.show()
```

```python
def true_LR(omega, tau):
    if omega*tau >= 1:
        N = int(omega*tau)
        return (sigma**2- psi**2/(4*omega))/N+ psi**2/(4*omega)*N
    return 1000


omega = 0.5
psi = 0.013304596290405304
sigma = 0.005236412970923553

tau_vals = np.linspace(0.000001,20,1000)
liquidity_risks = []
for tau in tqdm.tqdm(tau_vals):
    liquidity_risk_tau = get_liquidity_risk(tau, omega, psi, sigma, n_simulations = 1000)
    liquidity_risks.append(liquidity_risk_tau)

our_tau_opt = np.sqrt( (4*sigma**2)/(omega*(psi**2))- 1/(omega**2))

plt.plot(tau_vals, liquidity_risks, "b", label = "Agent-Based Simulation")
plt.plot(tau_vals, [2*sigma/(psi*np.sqrt(omega))]*len(tau_vals), 'r--')
plt.plot(tau_vals, sigma**2/(omega*tau_vals)+1/4*psi**2*tau_vals, 'g--', label = r"Fricke & Gerig Paper") #Theoretical result
plt.plot(tau_vals,[true_LR(omega, tau) for tau in tau_vals], 'r--', label = r"Our Result") #Theoretical result
plt.xlabel(r'Auction Duration $\tau$')
plt.ylabel(r'Liquidity Risk')
#plt.ylim([0, 5e-8])
plt.fill_between([2,4], 0, 0.0005, color = 'C1', alpha = 0.3, label= r"$\tau^{*}$")
plt.ylim([0, 0.0005])
plt.legend(loc = "best")
plt.show()
```
